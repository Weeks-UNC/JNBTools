# RNAvigate demonstration Notebook

This is a Jupyter notebook. It has a lot of nice features and is the gold
standard for reproducible data exploration and analysis. The idea that started
RNAvigate was to be able to make standard lab visualizations in Jupyter.

## Main elements of Jupyter notebooks

### 1. Markdown cells

This is a markdown cell. When you run this cell it will display as nicely
formatted HTML. I like to use these to summarize the main conclusions of an
analysis.

There are a few elements to markdown that are particularly useful.

2. To turn a line into a header add `#` for H1, `##` for H2, `###` for H3 etc.
2. For numbered lists, add `1.`, `2.`, etc. to the start of each line.
3. For bulleted lists, add `-` or `*` to the start of each line.
    * For nested lists, indent the line with 4 spaces.
4. For code blocks, wrap the code in `single ticks` (same key as tilde).
5. For multiline code blocks start and end with a line that only contains three
   single ticks

### 2. Code cells

This is where you put your python code.

### 3. Output cells

These are just the result of running code cells.


## Setting up the coding environment

You'll need to at least import RNAvigate. It can also be useful to import matplotlib.

```python
import rnavigate as rnav
import matplotlib.pyplot as plt
```

### Loading your data files

The next step is to load in data files. This is done with `rnav.Sample()`

For an actual analysis, this would be a good place to explain what data you have.

```python
example1 = rnav.Sample(
    sample="example1",
    shapemap="data/example1_rnasep_profile.txt",
    ct="data/RNaseP.ct",
    ss="data/RC_CRYSTAL_STRUCTURE.xrna",
    ringmap="data/example1-rnasep.corrs",
    pairmap="data/example1-rnasep-pairmap.txt",
    pairprob="data/rnasep.dp",
    log="data/example1_shapemapper_log.txt",
    dance_prefix="data/example1_rnasep",
    shapejump={
        "filename": "data/example-rnasep-deletions.txt",
        "fasta": "data/RNaseP-noSC.fasta"},
    pdb={
        "filename": "data/3dhs_Correct.pdb",
        "chain": "A"})

example2 =  rnav.Sample(
    sample="example2",
    shapemap="data/example2_rnasep_profile.txt",
    ct="data/RNaseP.ct",
    ss="data/RC_CRYSTAL_STRUCTURE.xrna",
    ringmap="data/example2-rnasep.corrs",
    pairmap="data/example2-rnasep-pairmap.txt",
    pairprob="data/rnasep.dp",
    log="data/example2_shapemapper_log.txt",
    dance_prefix="data/example2_rnasep",
    shapejump={
        "filename": "data/example-rnasep-deletions.txt",
        "fasta": "data/RNaseP-noSC.fasta"},
    pdb={
        "filename": "data/3dhs_Correct.pdb",
        "chain": "A"})

samples = [example1, example2]
```

## Make some plots

Now that you've loaded your data, you can make all the plots you want.

```python
plot = example1.plot_qc()
```

```python
plot = example1.plot_arcs(interactions="pairprob")
```

```python
plot = example2.plot_ss(
    interactions="ringmap",
    interactions_filter={
        "positive_only": True
    },
    profile="shapemap",
    colors="profile",
    apply_color_to="sequence")
```
