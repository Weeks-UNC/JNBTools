Recreation of DENV2 LowSS analysis
==================================

In this example, we will analyze SHAPE-MaP data from the Dengue 2 viral genome.
SHAPE-MaP of a long RNA such as this can be used to detect regions of well-
defined structures in which functional elements are over-represented. These
regions are defined by low SHAPE reactivity, indicating low local nucleotide
flexibility, and by low Shannon entropy, a measure of likelihood to form
alternative structures.

Import the RNAvigate python module
-----------------------------------

```python
import rnavigate as MaP
```

Provide file name locations for the data
----------------------------------------
In this case, the data is the result of running ShapeMapper, then SuperFold, on
the NGS sequencing reads.

```python
sample = MaP.Sample(sample="DENV2 Genome",
                    profile="DENV2_EX-1M7_MAP.map",
                    probs="DENV2_EX-Diff_BP-prob.dp",
                    ct="DENV2_EX-Diff_MFE.ct")
```

Plot the result of the low SHAPE, low Shannon entropy analysis
--------------------------------------------------------------

```python
MaP.LowSS(sample, region=(1, 10723//2));
```
