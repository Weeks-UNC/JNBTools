# Recreation of adenine riboswitch DANCE analysis

## Import StarMapper

```python
import starmapper as MaP
```

## Load data

```python
sample = MaP.Sample(sample="V.vul adeRS", dance_prefix="wt_0_deep")
sample.dance[0].data["ct"] = MaP.CT("ss", "wt_0_deep-0.nsd")
sample.dance[0].sample = "Translation OFF (42%)"
sample.dance[1].data["ct"] = MaP.CT("ss", "wt_0_deep-1.nsd")
sample.dance[1].sample = "Translation ON (58%)"
sample.dance[1].data["pdb"] = MaP.PDB("4tzx.pdb", chain="X")
```

## Make Arc plot for each DANCE component

```python
MaP.array_ap(sample.dance, ij="rings", cdAbove=15, positive_only=True,
             ij2="pairs", ij2_panel="top", colorbar=False);
```

## Make secondary structure drawings for each component

```python
for component in sample.dance:
    component.make_ss(ss="ct",ij="pairs", paired_only=True, ij2="rings",
                      ij2_filter={"cdAbove":15, "positive_only":True},
                      apply_color_to="sequence", colors="profile", colorbar=False);
```

## Draw 3D structure with RINGs

```python
sample.dance[1].make_mol(ij="rings", cdAbove=15, positive_only=True,
                         exclude_nts=[60], nt_color="grey");
```
