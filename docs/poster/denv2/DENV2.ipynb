Low SHAPE/Low Shannon entropy analysis of DENV2 genome
======================================================

This is a recreation of analysis and figure from Dethoff et al. 2018. Data files
were obtained from the authors.
- DENV2_EX_1M7_MAP.map
- DENV2_EX-Diff_BP-prob.dp
- DENV2_EX-Diff_MFE.ct

These files were produced by performing SHAPE-MaP, analyzing resulting sequencing
reads using Shapemapper, then producing a secondary structure model and base-
pairing probabilities using SuperFold.

In this example, I will analyze SHAPE-MaP data from the Dengue 2 viral genome.
SHAPE-MaP of a long RNA such as this can be used to detect regions of well-
defined structures in which functional elements are over-represented. These
regions are defined by low SHAPE reactivity, indicating low local nucleotide
flexibility, and by low Shannon entropy, a measure of likelihood to form
alternative structures. We call this a LowSS analysis.

```python
import rnavigate as rnav
```

Inputs
------

- sample
  - an arbitrary string that will serve as a label on plots
- shapemap
  - ShapeMapper2 output file, either _profile.txt or .map file
- pairprob
  - Superfold output file, .dp file providing basepairing probabilities
- ct
  - Superfold output file, .ct file providing an MFE structure

Notes:
- pairprob and ct can be generated from other structure prediction algorithms (Vienna RNAfold, NUPACK, RNAstructure, etc.)
- ct can also be provided in dotbracket, XRNA, VARNA, NSD, or CTE formats


```python
sample = rnav.Sample(
    sample="DENV2 Genome",
    shapemap="DENV2_EX-1M7_MAP.map",
    pairprob="DENV2_EX-Diff_BP-prob.dp",
    ct="DENV2_EX-Diff_MFE.ct")
```

Plot the result of the low SHAPE, low Shannon entropy analysis
--------------------------------------------------------------

First, I'll perform the low SHAPE, low Shannon entropy (LowSS) anal. By default, this will produce a full-length
plot. However, DENV2 is 10732 nt, and that's a bit unweildy, so I'm setting
`show=False` to turn off the plot for now.

```python
analysis = rnav.LowSS(sample, show=False)
```

I'll display only the first 2000 nt using the region argument:
`region=[start, end]`.

```python
analysis.plot_LowSS(region=[1, 2000])

# import matplotlib.pyplot as plt
# plt.savefig("denv2.svg")
```

We can also zoom in on one of the numbered regions above by providing that
number to the same region argument: `region=region_number`

```python
analysis.plot_LowSS(region=5)
```
