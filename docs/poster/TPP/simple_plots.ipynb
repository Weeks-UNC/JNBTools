```python
import rnavigate as rnav
```

```python
tpp = rnav.Sample(
    sample="TPP DMS-MaP",
    dmsmap="DMS_TPP_profile.txt",
    ringmap="DMS_TPP_rings.txt",
    ss="TPP-2GDI.nsd",
    pairprob="TPP-dms-bp.dp",
    pdb={"filepath":"2gdi.pdb", "chain":"X"})
```

```python
plot = tpp.plot_skyline(columns=["Modified_rate", "Untreated_rate"], region=[15,95])
plot.axes[0,0].get_xaxis().set_visible(False)
plot.axes[0,0].get_yaxis().set_visible(False)
plot.axes[0,0].get_legend().set_visible(False)
# plot.save("skyline.svg")
```

```python
plot = tpp.plot_shapemapper(plots=["profile"])
plot.axes[0,0].get_xaxis().set_visible(False)
plot.axes[0,0].get_yaxis().set_visible(False)
# plot.save("profile.svg")
```

```python
plot = tpp.plot_arcs(
    ct="ss",
    interactions="pairprob",
    title=False,
    seqbar=False,
    colorbar=False)
plot.axes[0,0].get_xaxis().set_visible(False)
# plot.save("arcplot.svg")
```

```python
plot = tpp.plot_ss(
    profile="dmsmap",
    colors="profile",
    apply_color_to="structure",
    title=False,
    colorbar=False,
    bp_style="conventional")
# plot.save("ssdiagram.svg")
```

```python
plot = tpp.plot_mol(
    interactions="ringmap",
    interactions_filter={
        "ct": "ss",
        "positive_only": True,
        "Statistic_ge": 90,
        "min_cd": 14},
    width=800,
    height=800,
    hide_cylinders=True,
    background_alpha=0,
    nt_color="grey",
    title=False,
    show=False)
plot.view.zoomTo({"chain":tpp.data["pdb"].chain})
plot.view.show()
```
