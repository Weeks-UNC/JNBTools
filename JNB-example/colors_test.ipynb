```python
import matplotlib.colors as mc
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.colors import ListedColormap

def view_colormap(cmap, ticks, values):
    cmap = plt.cm.get_cmap(cmap)
    colors = cmap(np.arange(cmap.N))
    
    fig, ax = plt.subplots(1, figsize=(6, 2))
    ax.imshow([colors], extent=[0, 10, 0, 1])
    ax.set_xticks(ticks)
    ax.set_xticklabels(values)
    ax.set_yticks([])
```

```python
cmap = ListedColormap([(80/255., 80/255., 80/255., 0.2), (0., 0., 243/255., 0.6), (30/255., 194/255., 1., 0.6)])
view_colormap(cmap, [1.75, 5, 8.25], ["All", "Primary", "Secondary"])
```

```python
view_colormap("turbo", [0, 10], ["Min", "Max"])
```
